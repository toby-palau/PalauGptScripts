```python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma
from langchain.document_loaders import PyPDFLoader, PDFMinerPDFasHTMLLoader, PyMuPDFLoader, AmazonTextractPDFLoader
from langchain.text_splitter import CharacterTextSplitter
from bs4 import BeautifulSoup
import re


loader = PyPDFLoader("./ESRS_E1_new.pdf")
# text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)

document = ""
for page in loader.load():
    match = re.search(r"Page \d+ of \d+", page.page_content)
    if match:
        document += page.page_content[match.end()+1:]
    else:
        document += page.page_content

document = document.replace("\n", "newlineplaceholder")
match = re.search(r"Table of content.*?(( )*?newlineplaceholder( )*?)+", document)
document = document[match.end():]

toc = []

while True:
    if len(toc) > 1:
        if re.search(f"^( *?)(( )*?newlineplaceholder( )*?)+?( *?){toc[0]}( *?newlineplaceholder)", document):
            break
    match = re.search(r"^( *?)(.+?)( +?)(\d+)( *?newlineplaceholder)", document)
    print(match.group(4), match.group(2))
    toc += ((match.group(2), match.group(4)))
    document = document[match.end():]

document = re.search(r"^(.+)")
```
